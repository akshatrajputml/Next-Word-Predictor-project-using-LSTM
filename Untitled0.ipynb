```python
import gradio as gr
import numpy as np
import pickle

from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Load model
model = load_model("/content/drive/MyDrive/AI   ML 2025/text reccomendation/ML MODEL/lstm_model.h5")

# Load tokenizer
with open("/content/drive/MyDrive/AI   ML 2025/text reccomendation/ML MODEL/tokenizer.pkl", "rb") as f:
    tokenizer = pickle.load(f)

# Load max length
with open("/content/drive/MyDrive/AI   ML 2025/text reccomendation/ML MODEL/max_len.pkl", "rb") as f:
    max_len = pickle.load(f)

# Reverse index
index_to_word = {}

for word, index in tokenizer.word_index.items():
    index_to_word[index] = word


# Prediction function
def predict_next_word(text):

    seq = tokenizer.texts_to_sequences([text])[0]

    seq = pad_sequences([seq], maxlen=max_len, padding='pre')

    prediction = model.predict(seq, verbose=0)

    predicted_word_index = np.argmax(prediction)

    return index_to_word.get(predicted_word_index, "Not Found")


# Gradio UI
app = gr.Interface(
    fn=predict_next_word,
    inputs=gr.Textbox(
        lines=2,
        placeholder="Enter your text here..."
    ),
    outputs="text",
    title="🤖 Next Word Predictor",
    description="Deep Learning NLP Model using LSTM"
)

app.launch(share=True)
```
